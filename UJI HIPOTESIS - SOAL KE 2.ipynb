## **UJI HIPOTHESIS 2:**

#### 2. Tagihan kesehatan dengan BMI diatas 25 lebih tinggi daripada tagihan kesehatan dengan BMI dibawah 25
  
      hipotesis yang akan kita uji adalah:
                              
$$H_0: \mu_{BMI>25} \geq \mu_{BMI<25}$$
$$H_1: \mu_{BMI>25} < \mu_{BMI<25}$$

```python
# load data
import pandas as pd
import numpy as np

# visualization
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure

from scipy.stats import t
from scipy.stats import ttest_ind

data = pd.read_csv('insurance.csv')
data = data.sample(100)
```

```python
#data BMI > 25
sample_upperBMI = data.loc[data.bmi > 25]
sample_upperBMI
```

```python
#data BMI< 25
sample_lowerBMI = data.loc[data.bmi<25]
sample_lowerBMI
```

```python
#Menghitung varian sample
np.var(sample_upperBMI.charges), np.var(sample_lowerBMI.charges)
```

```python
#perhitungan uji statistik t untuk uji rata rata 2 sampel menggunakan scipy.stats
stat, p = ttest_ind(a = sample_upperBMI.charges, b = sample_lowerBMI.charges, equal_var=False, alternative='less')

# Interpretasi Hasil
print('Statistics = %.4f, p = %.4f' % (stat, p)) 
```

```python
alpha = 0.05
# Pengambilan Keputusan
if p > alpha:
    print('Gagal tolak H0') 
else:
    print('Tolak H0')
```

Dari pengujian hipotesis diatas, kita mengambil kesimpulan bahwa tagihan kesehatan dengan BMI diatas 25 lebih tinggi daripada tagihan kesehatan dengan BMI dibawah 25
